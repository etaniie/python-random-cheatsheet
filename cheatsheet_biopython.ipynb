```python
# how to convert a string into a seq object.

DNA = 'AGTATAGGGCCCTTATATA'

print(f'Sequence stored in variable named DNA: {DNA} and its datatype is: {type(DNA)}')
```

```python
my_DNA_str = 'TATATATGGCCTAGC'
my_DNA_seq = Bio.Seq.Seq(my_DNA_str)

my_protein_str = 'QQQAWYTED'
my_protein_seq = Bio.Seq.Seq(my_protein_str)

print(my_DNA_seq)
print(my_protein_seq)
```

```python
# translation :

translation_DNA = my_DNA_seq.translate()
translation_DNA
```

```python
# reverse complement :

reverse_complement_DNA = my_DNA_seq.reverse_complement()
reverse_complement_DNA
```

```python
c_DNA = my_DNA_seq.complement()
c_DNA
```

```python
# task to calculate the GC_content of each sequence that is there in this file

# output : <header>: <GC_content>
```

```python
import Bio.SeqIO
from Bio.SeqUtils import gc_fraction
from Bio.SeqUtils import molecular_weight 
```

```python
file_content_bio = Bio.SeqIO.parse('c_elegans_100.fa', 'fasta')
for fasta in file_content_bio:
    id_fasta = fasta.id
    seq_fasta = fasta.seq

    GC_content = round(gc_fraction(seq_fasta), 2)
    mol_weight = round(molecular_weight(seq_fasta, 'DNA'), 2)
    print(f'{id_fasta:<10} : {GC_content:<12}  {mol_weight:<10}')
```

```python
write_fasta = open('fasta_GC_Molwt.txt', 'w')

file_content_bio = Bio.SeqIO.parse('c_elegans_100.fa', 'fasta')
for fasta in file_content_bio:
    id_fasta = fasta.id
    seq_fasta = fasta.seq

    GC_content = round(gc_fraction(seq_fasta), 2)
    mol_weight = round(molecular_weight(seq_fasta, 'DNA'), 2)
    write_fasta.write(f'{id_fasta}: {GC_content} {mol_weight}')

write_fasta.close()
```
