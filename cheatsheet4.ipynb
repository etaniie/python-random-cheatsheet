```python
# how a list can be created from a string

string_1 = 'python'

# this list will contain the individual characters of the above string as its element.

list_from_string_1 = list(string_1)
print(list_from_string_1)
print(len(list_from_string_1))
```

```python

tuple_from_string_1 = list(string_1)
print(tuple_from_string_1)
print(len(tuple_from_string_1))
```

```python
# .join()

# recover the string from the list

recovered_string_from_list = "".join(list_from_string_1)
print(f'recovered string from list: {recovered_string_from_list}')
```

```python
# recover the string from the tuple

recovered_string_from_tuple = "".join(tuple_from_string_1)
print(f'recovered string from tuple: {recovered_string_from_tuple}')
```

```python
## situation: 2  .split()
# .split() method never returns a tuple. It always returns a list.

string_1 = 'python'

list_2_from_string_1 = string_1.split() # both are same # string_1.split(" ")
print(list_2_from_string_1)

tuple_2_from_list_2 = tuple(list_2_from_string_1)
print(tuple_2_from_list_2)
```

```python
recovered_string_from_list_2 = "".join(list_2_from_string_1)
print(recovered_string_from_list_2)
recovered_string_from_tuple_2 = "".join(tuple_2_from_list_2)
print(recovered_string_from_tuple_2)
```

# summary

-If we know the index position either in a list/tuple, we can extract the corresponding value by listname[index]/tuplename[index]


-If we know the value from our list/tuple we can know the corresponding index by listname.index("value")/tuplename.index("value")

-there will be errors, if you try to access value or index which is does not exist in your list/tuple.

# Concatenation of a new list to an existing list

```python
list_genes = ['Myc', 'C9orf72', 'KRAS']
new_genes = ['NFKBIZ', 'BRCA1']
```

```python
# Approach I : create a new list without tempering the original list

# case I : Addition at the end 

list_genes_new_3 = list_genes + new_genes
print(list_genes_new_3)

# case II : Add somewhere in the middle 

list_genes_new_4 = list_genes[0:1] + new_genes + list_genes[1:]
print(list_genes_new_4)
```

```python
# Approach II : modify the original list

# case I : Addition at the end, apply .extend() rather than .append()

list_genes = ['Myc', 'C9orf72', 'KRAS']
list_genes.extend(new_genes)
print(list_genes)

# case II : add at a given index

list_genes = ['Myc', 'C9orf72', 'KRAS']
list_genes[1:1] = new_genes
print(list_genes)
```

```python
# Append vs Extend   # .append() will not work.

tuple_genes = ('Myc', 'C9orf72', 'KRAS')

new_gene = 'NFKBIZ'

tuple_genes_new = tuple_genes + (new_gene,)
tuple_genes_new
```

# sorting list and tuples

```python
# List sort

list_genes = ['C9orf72', 'Myc', 'KRAS', 'BRCA1','NFKBIZ', 'Myc']

# Approach I : sorted()

list_genes_sorted = sorted(list_genes)
print(list_genes)
print(list_genes_sorted)

print('*********************************************')

list_genes_sorted_desc = sorted(list_genes, reverse=True)
print(list_genes)
print(list_genes_sorted_desc)

# using a slicing approach 

print('*********************************************')
list_genes_sorted_desc_2 = list_genes_sorted[::-1] 
print(list_genes)
print(list_genes_sorted_desc_2)
```

```python
# Approach II : .sort()

list_genes_copy = list_genes.copy()
print(f'before appyling .sort(): {list_genes_copy}')

list_genes_copy.sort()
print(f'after applying .sort(): {list_genes_copy}')

print('*********************************************************')

list_genes_copy = list_genes.copy()
print(f'before appyling reverse: {list_genes_copy}')

list_genes_copy.sort(reverse=True)
print(f'after applying reverse: {list_genes_copy}')
```

# zip/unzip

```python
# zip

gene_list = ['Myc', 'C9orf72', 'KRAS', 'BRCA1']
expression_list = [-10, 0, 20, 35]

print(gene_list)
print(expression_list)
```

```python
gene_information = list(zip(gene_list, expression_list))
print(gene_information)
```

```python
# unzip

gene_information = [('Myc', -10), ('C9orf72', 0), ('KRAS', 20), ('BRCA1', 35)]
print(gene_information)
gene_list, expression_list = zip(*gene_information)
gene_list = list(gene_list)
expression_list = list(expression_list)

print(gene_list)
print(expression_list)
```
