```python
# how a list can be created from a string

string_1 = 'python'

# this list will contain the individual characters of the above string as its element.

list_from_string_1 = list(string_1)
print(list_from_string_1)
print(len(list_from_string_1))
```

```python

tuple_from_string_1 = list(string_1)
print(tuple_from_string_1)
print(len(tuple_from_string_1))
```

```python
# .join()

# recover the string from the list

recovered_string_from_list = "".join(list_from_string_1)
print(f'recovered string from list: {recovered_string_from_list}')
```

```python
# recover the string from the tuple

recovered_string_from_tuple = "".join(tuple_from_string_1)
print(f'recovered string from tuple: {recovered_string_from_tuple}')
```

```python
## situation: 2  .split()
# .split() method never returns a tuple. It always returns a list.

string_1 = 'python'

list_2_from_string_1 = string_1.split() # both are same # string_1.split(" ")
print(list_2_from_string_1)

tuple_2_from_list_2 = tuple(list_2_from_string_1)
print(tuple_2_from_list_2)
```

```python
recovered_string_from_list_2 = "".join(list_2_from_string_1)
print(recovered_string_from_list_2)
recovered_string_from_tuple_2 = "".join(tuple_2_from_list_2)
print(recovered_string_from_tuple_2)
```

```python
string_2 = 'i love python'

list_3_from_string_2 = string_2.split(" ")
print(list_3_from_string_2)
```

```python
tuple_3_from_list_3 = tuple(list_3_from_string_2)
print(tuple_3_from_list_3)
```

```python
recovered_string_from_list_3 = "".join(list_3_from_string_2)
print(recovered_string_from_list_3)

recovered_string_from_tuple_3 = "".join(tuple_3_from_list_3)
print(recovered_string_from_tuple_3)
```

# summary

-If we know the index position either in a list/tuple, we can extract the corresponding value by listname[index]/tuplename[index]


-If we know the value from our list/tuple we can know the corresponding index by listname.index("value")/tuplename.index("value")

-there will be errors, if you try to access value or index which is does not exist in your list/tuple.

```python
# CONCATENATION

list_genes = ['Myc', 'C9orf72', 'KRAS'] # 'NFKBIZ' 'BRCA1', 'BRCA2', 'Her2'
```

```python
list_genes
```

```python
# concatenation of a single element to an existing list

new_gene = 'NFKBIZ'

# Goal: concatenate this new_gene to list_genes 

# Approach 1 : create a new list

# this one is a case where we added a new element at the end of the existing list
list_genes_new = list_genes + [new_gene] 
print(list_genes_new)
```

```python
# this one is a case where we added a new element at a given index of existing list

list_genes_new_2 = list_genes[0:1] + [new_gene] + list_genes[1:]

print(list_genes_new_2)
```

```python
# Approach 2 : modify the original list itself

# method used here is .append()

list_genes.append(new_gene)
```
