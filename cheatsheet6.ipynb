# Decision making in python

```python
# logical operation

# ==

print(2 == 2)
print(2 == 3)
print('********')

# <

print(2 < 3)
print(2 < 1)
print('********')

# <=

print(2 <= 2)
print(2 <= 3)
print(2 <= 1)
print('********')

# > 

print(3 > 2)
print(3 > 4)
print('********')

# >=

print(8 >= 8)
print(8 >= 7)
print(8 >= 9)
print('********')

# !=

print(2 != 3)
print(2 != 2)
print('********')
```

```python
# Strings 

print("ATTCCG".startswith('ATT'))
print("ATTCCG".startswith('ATG'))
print("ATTCCG".endswith('ATC'))
print("ATTCCG".endswith('CCG'))
```

```python
# Membership 

gene_name = ['Myc', 'C9orf72', 'NFKBIZ']

print('Myc' in gene_name)
print('Myc' not in gene_name)

print('BRCA1' in gene_name)
print('BRCA1' not in gene_name)
```

```python
# OR

print((1 < 2) or (2 > 3))
print((1 < 2) or (2 < 3))
print((1 > 2) or (2 > 3))

print((1 < 2) or (2 > 3) or (4 > 5))
```

```python
# and

print((1 < 2) and (2 > 3))
print((1 < 2) and (2 < 3))
print((1 > 2) and (2 > 3))

print((1 < 2) and (2 > 3) or (4 > 5))
```

```python
# not

print(2 < 3)
print(not 2 < 3)
```

# If, Else and Elif

```python
# Activity I

# mark the gene which are highly expressed and other which are low expressed 

gene_expression = [20, 10, 2, 30, 55, 120, 150, 200]

# output : if expression value is < 20: mark it 'low expression'; else: mark 'high expression'

for expression in gene_expression:
    if expression < 20:
        print(f"Gene with expression value of {expression} is considered to have 'low expression'")
    else:
        print(f"Gene with expression value of {expression} is considered to have 'high expression'")
```

```python
for expression in gene_expression:
    if expression < 20:
        print(f"Gene with expression value of {expression} is considered to have 'low expression'")
    elif (expression >= 20) and (expression < 100):
        print(f"Gene with expression value of {expression} is considered to have 'moderate expression'")
    else:
        print(f"Gene with expression value of {expression} is considered to have 'high expression'")
```

# Activity II: we have a DNA sequence and we have to verify whether
# this sequence is a valid coding sequence
# it should start with a start codon
# the sequence length must be a multiple of 3

```python
# sequence = 'ATGCTTAGGCTA'
sequence = "ACGCTTAGGCTA"

start_codon = 'ATG'

if sequence.startswith(start_codon):
    if len(sequence) % 3 ==0:
        print('Sequence start with a start codon and it is a valid coding sequence')
    else: 
        print('Although the sequence start with start codon, It is not a valid coding sequence')
else:
    print('Sequence does not start with a start codon.')
```
