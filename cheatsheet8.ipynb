# Creating Dictionary 

```python
# case 1 : empty dictionary

# method I : 

dict_1 = {}

# method II : 

dict_2 = dict()

print(f'dict_1: {dict_1} and its datatype:{type(dict_1)}' )
print(f'dict_2: {dict_2} and its datatype:{type(dict_2)}' )
```

```python
# case 2 : non-empty dictionary

# method I :

dict_3 = {'gene name' : 'TP53', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome'}

print(f'dict_3: {dict_3} and its datatype:{type(dict_3)}' )
```

```python
# method II :

dict_4 = dict(gene_name = 'TP53', location = '17p13.1', disease = 'Li-Fraumeni Syndrome')
print(f'dict_4: {dict_4} and its datatype: {type(dict_4)}')
```

```python
# method III : Dictionary from a pair lists with equal length 

list_a = ['gene_name', 'location', 'disease']
list_b = ['TP53', '17P13.1', 'Li-Fraumeni Syndrome']

dict_5 = dict(list(zip(list_a, list_b)))
print(f'dict_5: {dict_5} and its datatype:{type(dict_5)}')
```

```python
dict_6 = {'gene name' : 'TP53', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome',
         'expression' : 120}
print(f'dict_6: {dict_6} and its datatype:{type(dict_6)}')
```

```python
# keys are supposed to be unique

dict_8 = {'gene_name' : 'MYC', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome',
         'expression' : 120, 'gene_name' : 'TP53'}
dict_8
```

```python
# values can be non-unique 

dict_9 = {'gene_name_1' : 'MYC', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome',
         'expression' : 120, 'gene_name_2' : 'TP53', 'gene_name_3' : 'MYC'}
dict_9
```

```python
# lenght of a dictionary: the number of element (pair of key and value)

# len()

print(f'lenght of this dictionary is {len(dict_9)}')
```

```python
# copy of a dictionary

# case I : copy the value

dict_10 = dict_9.copy()

# case II : copy it by address

dict_11 = dict_9
```

```python
print(f'content of dict_9: {dict_9} and its id: {id(dict_9)}')
print(f'content of dict_9: {dict_10} and its id: {id(dict_10)}')
print(f'content of dict_9: {dict_11} and its id: {id(dict_11)}')
```

```python
# how to extract the keys and value 

# values can be non-unique

dict_9 = {'gene_name_1' : 'MYC', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome',
         'expression' : 120, 'gene_name_2' : 'TP53', 'gene_name_3' : 'MYC'}
dict_9_keys = dict_9.keys()
dict_9_values = dict_9.values()

print(list(dict_9_keys))
print(list(dict_9_values))
```

```python
# value corresponding to a given key which is 'location'

print(dict_9['location'])
print(dict_9['expression'])
```

```python
# put in some key which does not exist in our dictionary 

print(dict_9['expression_1'])
```

```python
# method 2 : which is more safe in order to get away with these kinds of keyerror

print(dict_9.get('expression_1', 'key does not exist'))
print(dict_9.get('expression', 'key does not exist'))
```

# adding / updating an existing dictionary

```python
dict_3 = {'gene_name': 'TP53', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome'}

print(f'before: {dict_3}')
# i want to add a new key: value pair. 'expression' : 120

print(dict_3['gene_name'])
dict_3['expression'] = 120
print(f'after: {dict_3}')
# dict_3['expression'] = 120
```

```python
# update the value of an existing key
dict_3['expression'] = 150
print(f'after update: {dict_3}')
```

```python
# common confusion, that many people experience when they are working with list and dict

list_1 = ['gene_name', 'location', 'disease']
print(f'before:{list_1}')
# update an existing element. location -> location
list_1[1] = 'Location'
print(f'after: {list_1}')
```

```python
# case 2: where i want to add a new element
# list_1[3] = 'expression' ### wrong approach
list_1.append('expression')
print(list_1)
```

# how to iterate over the elements of a dictionary 

```python
# we can iterate in three different manners
## over the keys only
## over the values only 
## over (key,value) pair
```

```python
print(dict_3)
```

```python
# iterate over the keys:
for key in dict_3.keys():
    print(key)
```

```python
# iterate over the values:
for value in dict_3.values():
    print(value)
```

```python
# iterate over both key: value 
for key, value in dict_3.items():
    print(f'{key}: {value}')
```

# delete contents of a dictionary: pop()

## to delete the individual elements of a dictionary;
~ pop()
~ popitem()
~ del

## to delete the complete dictionary

~ clear
~ del

```python
dict_3 = {'gene_name': 'TP53', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome'}
print(dict_3)
```

```python
# pop() : you apply this in a situation where you want to delete a (key:value) pair but you also 
# want to store the value corresponding to the key which has been deleted.

dict_3_copy = dict_3.copy()
print(f'before: {dict_3_copy}')
value_delete = dict_3_copy.pop('location')
print(f'value_delete: {value_delete}')
print(f'after: {dict_3_copy}')
```

```python
# popitem() : you apply this in a situation where you want to delete a (key: value) pair but you also
# want to store both the key and the value which was deleted

dict_3_copy = dict_3.copy()

print(f'before: {dict_3_copy}')
key_deleted, value_deleted = dict_3_copy.popitem()
print(f'key_deleted: {key_deleted}, value_deleted: {value_deleted}')
print(f'after: {dict_3_copy}')
```

```python
# del : when we want to delete a key, value pair but we don't bother storing the content which 
# was deleted

dict_3_copy = dict_3.copy()
print(f'before: {dict_3_copy}')

del dict_3_copy['location']
print(f'after: {dict_3_copy}')
```

```python
# case where we want to complete delete the dictionary 
# clear 

dict_3_copy = dict_3.copy()
print(f'before: {dict_3_copy}')

dict_3_copy.clear()
print(f'after: {dict_3_copy}')
```

```python
# del

dict_3_copy = dict_3.copy()
print(f'before: {dict_3_copy}')

del dict_3_copy # this is wrong
print(f'after: {dict_3_copy}')
```

```python
# creating new dictionary (rather than modifying an existing one) to delete few key: value pair

dict_10 = {'gene_name': 'TP53', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome', 'expression' : 200}
print(f'dict_10: {dict_10}')

# goal is to create a new dict: dict_11 and I want its content to contain only this 'gene_name': 'TP53' and
# 'disease' : 'Li-Fraumeni Syndrome'

keys_to_delete = ['location', 'expression']
dict_11 = {key: value for key, value in dict_10.items() if key not in keys_to_delete}
print(f'dict_11: {dict_11}')
```
