# List and tuple interconversion

```python
# list to tuple 
```

```python
list_1 = [1, 'python', 4.5, True]

print(list_1)
```

```python
print(type(list_1))
```

```python
tuple_1 = tuple(list_1)

print(tuple_1)
print(type(tuple_1))
```

```python
# tuple to list
```

```python
tuple_2 = (1, 'python', 4.5, True)

print(tuple_1)
print(type(tuple_1))
```

```python
list_2 = list(tuple_2)

print(list_2)
print(type(list_2))
```

```python
# how to calculate the length of a list as well as length of a tuple
# how they differ from length of a string
```

```python
list_3 = ['python', 12, 5.6, 'Molecular Biology']

list_3_length = len(list_3)
list_3_length
```

```python
tuple_3 = ('python', 12, 5.6, 'Molecular Biology')

tuple_3_length = len(tuple_3)
tuple_3_length
```

```python
# string datatype

string_1 = 'I am going to use python for my molecular biology course'

string_1_length = len(string_1)
string_1_length
```
