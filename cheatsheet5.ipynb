# Loops in python

```python
# case 1 : we will use a one-liner for loop to create a list

a = range(0, 6)

print(list(a))
# our goal is to make a list where I want the individual element of the range object to be raised to the power of 3.

list_number = [num ** 3 for num in a]
print(list_number)
```

```python
# one liner for loop has multiple application in list operation

## Activity I

# input 
names_student_python_class = ['Ana', 'Peter', 'Louis']

# task is to create a list of their email id. The pattern of mail id is: first_name@gmail.com
# e.g. name is 'Ana', email id will be ana@gmail.com

email_student_python_class = [x.lower() + '@gmail.com' for x in names_student_python_class]

print(email_student_python_class)
```

```python
## Activity II

names_student_python_class = ['Ana', 'Peter', 'Louis', 'John']
index_information = [1,3] # these students are missing the class

# goal : to filter the student corresponting to the index that we have 

filtered_names_student_python_class = [names_student_python_class[i] for i in index_information]
print(filtered_names_student_python_class)
```

# Conventional 'for' loop 

```python
# Syntax for the 'for' loop

#Input

proteins_list = ["Insulin", "Actin", "Fibrinogen"]

# Output:
# Insulin is a protein
# Actin is a protein
# Fibrinogen is a protein
```

```python
for protein in proteins_list: 
    print(protein + " is a Protein") 
    print("************************")
```

```python
# For loop can iterate over other datatypes as well.

for i in range(6):
    print(i) 
    print('\n')
```

```python
for i in range(6):
    print(i) 
print('\n')
```

```python
for letter in "I love python":
    print(letter)
```

```python
# Inputs:

gene_list = ['Myc', 'NFKBIZ', 'BRCA1']
expression_gene = [-20, 80, 10]

# Output: 
  # Gene 'Myc' shows an expression of -20
  # Gene 'NFKBIZ' shows an expression of 80
  # Gene 'BRCA1' shows an expression of 10

# Elaborate that sometimes it is better to iterate over the index of a list rather than indexing over its value

for i in range(len(gene_list)): # range(3)
    print(f'Gene{gene_list[i]} shows an expression of {expression_gene[i]}') 
    print('*************************************')
```

```python
for gene in gene_list:
    print(gene)
```

```python
for i in range(len(gene_list)):
    print(gene_list[i])
```

# Nested for loops:

```python
# loops within loops
```

```python
# Input list

dna_sequence = [['A', 'T', 'C'], 
                ['A', 'A', 'T', 'T', 'G'],
                ['C', 'A'],
                ['T', 'T', 'G', 'G', 'A', 'C']]

# Goal is to make a list of the length of the nested lists

# e.g. output will be [3, 5, 2, 6]
# Constraint: we cannot use the len() function

for row in dna_sequence:
    print(row)
    break 
```

```python
length_dna_sequence = []

for row in dna_sequence:
    count = 0
    for base in row:
        count = count + 1
        length_dna_sequence.append(count)

print(f'Length of nested-list with the main list is: {length_dna_sequence}')
```
