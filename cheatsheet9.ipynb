```python
# we can convert this piece of code into a function
# defining a function

def do_reverse_complement(input_DNA):
    rev_com_DNA = input_DNA.replace("A", "t")
    rev_com_DNA = rev_com_DNA.replace("T", "a")
    rev_com_DNA = rev_com_DNA.replace("C", "g")
    rev_com_DNA = rev_com_DNA.replace("G", "c")

    rev_com_DNA = rev_com_DNA.upper()[::-1]
    return rev_com_DNA    
```

```python
# function call 
input_DNA_1 = "ATGTGTAAGGATGTTAG"
input_DNA_2 = "AATGTAGGGGATGATGAAATTA"

rev_complement_DNA_1 = do_reverse_complement(input_DNA_1)
rev_complement_DNA_2 = do_reverse_complement(input_DNA_2)

print(rev_complement_DNA_1)
print(rev_complement_DNA_2)
```

```python
# function with no return value

def do_reverse_complement_no_return(input_DNA):
    rev_com_DNA = input_DNA.replace("A", "t")
    rev_com_DNA = rev_com_DNA.replace("T", "a")
    rev_com_DNA = rev_com_DNA.replace("C", "g")
    rev_com_DNA = rev_com_DNA.replace("G", "c")

    rev_com_DNA = rev_com_DNA.upper()[::-1]

    print(rev_com_DNA)
```

```python
# function call 
input_DNA_1 = "ATGTGTAAGGATGTTAG"
input_DNA_2 = "AATGTAGGGGATGATGAAATTA"

do_reverse_complement_no_return(input_DNA_1)
do_reverse_complement_no_return(input_DNA_2)
```

```python
# where i will pass two DNA sequence and we have to print their GC content upto 2 decimals and compare them 

def calculate_and_compare_GC_content(DNA_seq1, DNA_seq2, sig_dec):
    comparison_result = ''

    GC_content_DNA_1 = gc_fraction(DNA_seq1)
    GC_content_DNA_1 = round(GC_content_DNA_1, sig_dec)

    GC_content_DNA_2 = gc_fraction(DNA_seq2)
    GC_content_DNA_2 = round(GC_content_DNA_2, sig_dec)

    if GC_content_DNA_1 > GC_content_DNA_2:
        comparison_result = 'greater'
    
    else:
        comparison_result = 'less'

    return GC_content_DNA_1, GC_content_DNA_2, comparison_result
```

```python
DNA_seq1 = generate_random_dna() 
DNA_seq2 = generate_random_dna()
sig_dec = 2

GC_content_DNA_1, GC_content_DNA_2, comparison_message = calculate_and_compare_GC_content(DNA_seq1, DNA_seq2, sig_dec)

print(f'{GC_content_DNA_1:<8}{GC_content_DNA_2:<8}{"DNA1 content is" + comparison_message + "than DNA2":<20}')
```

# dynamic number of arguments in function definition

1st set of experiment:
   - gene expression: 2 times; averaged the value

2nd set of experiment;
   - gene expression: 3 times, averaged the value

3th set of experiment;
   - gene expression: 5 times, averaged the value

contraint: we have to make a single function, which takes the value of gene expression received in each experiment.
   - for e.g. 
   - if the experiment was repeated 2 times, function should be able to hand a scenario of average_gene_expression(a, b)
   - if the experiment was repeated 3 times, the function should handle the scenario of average_gene_expression(a,b,c)

```python
def average_gene_expression(*args):
    sum_expression = 0
    count_experiment = 0
    for expression_value in args:
        # sum_expression = sum_expression + expression_value 
        sum_expression += expression_value
        count_experiment += 1

    average_expression = sum_expression/count_experiment

    return round(average_expression, 2)
```

```python
# case 1 : we have two repeatition of the experiment 

average_value_expression = average_gene_expression(120, 140)
print(average_value_expression)
```

```python
# case 2 : we have three repeatition of the experiment

average_value_expression = average_gene_expression(120, 140, 125)
print(average_value_expression)
```

```python
# case 3 : we have ten repeatition of the experiment

average_value_expression = average_gene_expression(120, 140, 125, 100, 122, 120, 90, 144, 125, 136)
print(average_value_expression)
```

```python
def average_gene_expression_keyword(sign_dec=2, **kwargs):
    sum_expression = 0
    count_experiment = 0
    for id, expression_value in kwargs.items():
        # sum_expression = sum_expression + expression_value 
        sum_expression += expression_value
        count_experiment += 1

    average_expression = sum_expression/count_experiment

    return round(average_expression, 2)
```
