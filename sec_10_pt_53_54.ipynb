# Creating Dictionary 

```python
# case 1 : empty dictionary

# method I : 

dict_1 = {}

# method II : 

dict_2 = dict()

print(f'dict_1: {dict_1} and its datatype:{type(dict_1)}' )
print(f'dict_2: {dict_2} and its datatype:{type(dict_2)}' )
```

```python
# case 2 : non-empty dictionary

# method I :

dict_3 = {'gene name' : 'TP53', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome'}

print(f'dict_3: {dict_3} and its datatype:{type(dict_3)}' )
```

```python
# method II :

dict_4 = dict(gene_name = 'TP53', location = '17p13.1', disease = 'Li-Fraumeni Syndrome')
print(f'dict_4: {dict_4} and its datatype: {type(dict_4)}')
```

```python
# method III : Dictionary from a pair lists with equal length 

list_a = ['gene_name', 'location', 'disease']
list_b = ['TP53', '17P13.1', 'Li-Fraumeni Syndrome']

dict_5 = dict(list(zip(list_a, list_b)))
print(f'dict_5: {dict_5} and its datatype:{type(dict_5)}')
```

```python
dict_6 = {'gene name' : 'TP53', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome',
         'expression' : 120}
print(f'dict_6: {dict_6} and its datatype:{type(dict_6)}')
```

```python
dict_7 = {'gene name' : 'TP53', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome',
         'expression' : (120, 150, 180)}
dict_7
```

```python
# keys are supposed to be unique

dict_8 = {'gene_name' : 'MYC', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome',
         'expression' : 120, 'gene_name' : 'TP53'}
dict_8
```

```python
# values can be non-unique 

dict_9 = {'gene_name_1' : 'MYC', 'location' : '17p13.1', 'disease' : 'Li-Fraumeni Syndrome',
         'expression' : 120, 'gene_name_2' : 'TP53', 'gene_name_3' : 'MYC'}
dict_9
```
