# working with files

- .read(): it will retrieve the complete content of the file as a single string
- .readline(): this gives you an iterable object, where you can get a single line as a string everytime you do a .readline()
- .readlines(): this will provide you with a list, where every line of your file will become an element of this list.

```python
## new approach to opening a file?

# old approach: need to close the file_handle explicitly 
# read
pathname_fasta = r'C:\Users\eceka\Downloads\MYC_datasets\ncbi_dataset\data\gene.fna'

file_handle = open(pathname_fasta, 'r')

file_content = file_handle.read()
file_handle.close()

# new approach: no need the close the file_handle explicitly 

pathname_fasta = r'C:\Users\eceka\Downloads\MYC_datasets\ncbi_dataset\data\gene.fna'

# opening a file in read mode 'r'
# r: read, w: write, a: append
with open(pathname_fasta, 'r') as file_handle:
     file_content = file_handle.read()

print(file_content[:200])

```

```python
file_content[:200]
```

```python
file_content = file_content.replace('\n', '')
file_content[:200]
```

```python
file_content = file_content.replace('>NC_000008.11:127735434-127742951 MYC [organism=Homo sapiens] [GeneID=4609] [chromosome=8]' , '')
file_content[:200]
```

```python
#len() 

file_content_length = len(file_content)
print(f'The length of myc gene from this file was found to be {file_content_length}')
```

```python
#gc_content = .count()

gc_content = file_content.count('G') + file_content.count('C') 
gc_content = round((gc_content/file_content_length)*100, 2)
print(f'The GC content of myc gene from this file was found to be {gc_content}')
```
